# PSNR at Average or Box Blur

```python
original = cv2.imread(r'C:\Users\dvmsh\Desktop\Study mode\SRGAN Codes\Metrices code\data\HR_images\img5.jpg')
avg_blur_img = cv2.imread(r'C:\Users\dvmsh\Desktop\Study mode\SRGAN Codes\Metrices code\data\LR_images\img5_avg_blur.jpg',1)
```

```python
def compute_psnr(img1, img2):
    img1 = img1.astype(np.float64)
    img2 = img2.astype(np.float64) 
    mse = np.mean((img1 - img2) ** 2)
    if mse == 0:
        return "Same Image"
    
    return 20 * math.log10((255)/math.sqrt(mse))
    
```

```python
psnr=compute_psnr(original, avg_blur_img)
print(psnr)
```

# PSNR at Gaussian Blurr
